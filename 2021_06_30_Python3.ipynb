```python
#Euclidean distance
#https://www.cut-the-knot.org/pythagoras/DistanceFormula.shtml
import numpy as np

x1=np.array([2,-1])
y1=np.array([-2,2])

euclidean_distance=np.sqrt(np.sum((x1[0]-y1[0])**2+(x1[1]-y1[1])**2))
#euclidean_distance=np.sqrt(np.sum(np.power((x1-y1),2)))
print(euclidean_distance)
```

```python
#Euclidean distance
#https://www.cut-the-knot.org/pythagoras/DistanceFormula.shtml
import numpy as np

x1=np.array([2,-1,2,4])
y1=np.array([-2,2,1,3])
euclidean_distance=np.sqrt(np.sum(np.power((x1-y1),2)))
print(euclidean_distance)
```

```python
# Manhattan distance
#https://en.wikipedia.org/wiki/Taxicab_geometry
import numpy as np

x1=np.array([2,-1])
y1=np.array([-2,2])

manhattan_distance=sum(abs(x1-y1))
print(manhattan_distance)
```

```python
# cosine similarity
#https://www.oreilly.com/library/view/statistics-for-machine/9781788295758/eb9cd609-e44a-40a2-9c3a-f16fc4f5289a.xhtml
#https://en.wikipedia.org/wiki/Cosine_similarity
from numpy import dot
from numpy.linalg import norm

x1=np.array([2,-1])
y1=np.array([-2,2])

cosine_similarity = dot(x1, y1)/(norm(x1)*norm(y1))
print(cosine_similarity)
```

```python
#python package for distance
from scipy.spatial import distance
#https://docs.scipy.org/doc/scipy/reference/spatial.distance.html

x1=np.array([2,-1])
y1=np.array([-2,2])

euclidean_distance=distance.euclidean(x1,y1)
print("euclidean_distance:",euclidean_distance)
cosine_similarity=1-distance.cosine(x1,y1)
print("cosine_similarity:",cosine_similarity)
manhattan_distance=distance.cityblock(x1,y1)
print("manhattan_distance:",manhattan_distance)
```

```python
#k-means clustering

#https://www.analyticsvidhya.com/blog/2021/04/k-means-clustering-simplified-in-python/
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

data1=np.random.randn(100,2)
data2=np.random.randn(100,2)+4
data_row_bind=np.vstack((data1,data2))

plt.scatter(data_row_bind[:,0],data_row_bind[:,1])
plt.show()

kmeans_result = KMeans(n_clusters=2, random_state=0).fit(data_row_bind)
plt.scatter(data_row_bind[:,0],data_row_bind[:,1],c=kmeans_result.labels_)
plt.show()
```
