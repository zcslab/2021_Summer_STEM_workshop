```python
#Set
# like list or tuple
my_set=set()#empty set, 
#Don't use my_set={} to empty a set 
#only one way to empty a set, not like tuple or list 
print("empty set:",my_set)

my_set={1,2,3,4}# or my_set=set([1,2,3,4]) or my_set=set((1,2,3,4))
print(my_set)
print(my_set(0))#Error! 
print(my_set(2:4))#Error!
```

```python
#Set
#Adding
print("Adding")
my_set={1,2,3}
print(my_set)
my_set.add(4)
print(my_set)

#updating, another way to add element to a set
print("Updating")
my_set={1,2,3}
print(my_set)
my_set.update({5})
print(my_set)

#deleting
print("deleting")
my_set={1,2,3}
print(my_set)
my_set.remove(2)
print(my_set)

#discarding, another way to remove element
print("discard")
my_set={1,2,3}
print(my_set)
my_set.discard(2)
print(my_set)

#pop, remove an element.
print("pop")
my_set={1,2,3}
print(my_set)
my_set.pop()
print(my_set)

#clear the set
print("clearing")
my_set={1,2,3}
print(my_set)
my_set.clear()# empty the set
print(my_set)
```

```python
# Set
#Operations between two sets
set1=set("abcd")
set2=set("abce")
print(set1)
print(set2)

print(set1-set2)# elements in set1, not in set2
print(set1 | set2)# elements in set1 or set2
print(set1 & set2)#elements in set1 and set2
print(set1 ^ set2)#elements not in set1 and not in set2
```

```python
# Function
#Fibonacci series: 0,1,1,2,3,5,8
#def function_name(inputs):
#   codes

def Fibonacci(n):
    a, b = 0, 1
    while b < n:
        print(b,end=',')# the outputs in one line
        #print(b)
        a, b = b, a+b
#how to call the function
Fibonacci(10)# don't need print because print included in the function.
```

```python
# Function
# list adding function

def list_adding(my_list):
    for i in range(5,9):
        my_list.append(i)
    return(my_list)# quit the function

# call the function
my_list=[1,2,3]
print(my_list)
list_adding(my_list)
print(my_list)
# pay attention, the list my_list got changed.  
```

```python
#Function
# list adding function, what if we want to keep both of them
def list_adding(my_list):
    for i in range(5,9):
        my_list.append(i)
    return(my_list)

# call the function
my_list=[1,2,3]
print(my_list)
my_new_list=list_adding(my_list.copy())# copy List my_list to function "list_adding"
print(my_list)
print(my_new_list)
# pay attention, the list my_list not got changed. 
```
